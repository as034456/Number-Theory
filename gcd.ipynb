```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

def find_order(a, n):
    if gcd(a, n) != 1:
        return None  # a and n must be relatively prime
    order = 1
    current = a % n
    while current != 1:
        current = (current * a) % n
        order += 1
    return order

def main():
    # Get user input
    a = int(input("Enter value for a (must be relatively prime to n): "))
    n = int(input("Enter value for n: "))
    
    # Check if a and n are relatively prime
    if gcd(a, n) != 1:
        print(f"{a} and {n} are not relatively prime. Please enter valid values.")
        return

    # Find and display the order of a modulo n
    order_a = find_order(a, n)
    print(f"Order of {a} modulo {n}: {order_a}")

if __name__ == "__main__":
    main()
```

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

def euler_totient(n):
    result = n
    p = 2
    while p * p <= n:
        if n % p == 0:
            while n % p == 0:
                n //= p
            result -= result // p
        p += 1
    if n > 1:
        result -= result // n
    return result

def is_primitive_root(g, n):
    if gcd(g, n) != 1:
        return False
    phi_n = euler_totient(n)
    for k in range(1, phi_n):
        if phi_n % k == 0 and pow(g, k, n) == 1:
            return False
    return True

def find_primitive_roots(n):
    roots = []
    for g in range(2, n):
        if is_primitive_root(g, n):
            roots.append(g)
    return roots

def construct_indices_table(g, n):
    table = {}
    current = 1
    for i in range(n - 1):
        table[current] = i
        current = (current * g) % n
    return table

def mod_inverse(a, m):
    m0, x0, x1 = m, 0, 1
    if m == 1:
        return 0
    while a > 1:
        q = a // m
        m, a = a % m, m
        x0, x1 = x1 - q * x0, x0
    if x1 < 0:
        x1 += m0
    return x1

def solve_congruence(a, c, n, g, indices_table):
    if a not in indices_table or c not in indices_table:
        return None  # a or c not found in indices table
    alpha = indices_table[a]
    gamma = indices_table[c]
    phi_n = euler_totient(n)
    # Solve alpha * x ≡ gamma (mod phi(n))
    # x ≡ gamma * alpha^(-1) (mod phi(n))
    alpha_inv = mod_inverse(alpha, phi_n)
    x = (gamma * alpha_inv) % phi_n
    return x

def main():
    n = int(input("Enter value for n: "))
    
    primitive_roots = find_primitive_roots(n)
    if not primitive_roots:
        print(f"No primitive roots found for {n}")
        return
    
    print(f"Primitive roots of {n}: {primitive_roots}")
    
    g = int(input(f"Enter a primitive root from the list above: "))
    
    indices_table = construct_indices_table(g, n)
    print(f"Indices table for primitive root {g} modulo {n}:")
    for key, value in indices_table.items():
        print(f"{key}: {value}")
    
    a = int(input("Enter value for a (must be in the range [1, {n-1}]): "))
    c = int(input("Enter value for c (must be in the range [1, {n-1}]): "))
    
    if a not in indices_table or c not in indices_table:
        print(f"Either {a} or {c} is not in the indices table. Please enter valid values.")
        return
    
    x = solve_congruence(a, c, n, g, indices_table)
    if x is None:
        print("No solution found.")
    else:
        print(f"The solution to {a}^x ≡ {c} (mod {n}) is x = {x}")

if __name__ == "__main__":
    main()
```
